```python
def extract_entities(text):

    wanted = ['ADJ','NOUN','PROPN','ADV']
    nlp = en_core_web_sm.load()
    preprocess_text(text, [lemmatize_word])
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents] #if ent.pos_ in wanted ] # TODO
    return entities
```

```python
text = "Google appeared in 1998 thanks to the efforts of Larry Page and Sergey Brin.\
 Initially, a student and a post-graduate of the Stanford University developed an \
 algorithm that ranked sites contingent on the number of backlinks for every URL. \
 By the end of the year, Google had indexed 60 million pages and received its first positive feedback."
text = preprocess_text(text, [lemmatize_word])
```

```python
article = nlp(text)
```

```python
# @title Working with the example text
entities_ent = [] # all entities found with spacy's entity finder
all = [] # all tokens with POS-labels
entities_pos = [] # just the POS we want (defined in wanted)
wanted = ['ADJ','NOUN','PROPN','ADV']

for ent in article.ents:
  entities_ent.append(ent.text)
  #print("Text: ",ent.text,"\tLabel: ", ent.label_)
print(entities_ent)
for ent in article:
  all.append((ent.text,ent.pos_))
  if ent.pos_ in wanted:
    entities_pos.append(ent.text)

print(all)
print(entities_pos)
```

```python
import nltk
from nltk import word_tokenize, pos_tag
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
for sent in nltk.sent_tokenize(text):
  for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent))):
    if(hasattr(chunk, 'label')):
      print(chunk.label(), ' '.join(c[0] for c in chunk))
```

```python
# @title Evaluation of Entity-recognition - Spacy-result as benchmark - several occurences are not taken into account
def compare(ground_truth, res):
  inters = set(ground_truth).intersection(res)
  print(inters)
  return (len(inters)/max(len(ground_truth),len(res)))
```
